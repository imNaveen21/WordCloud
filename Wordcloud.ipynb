```python
import wikipedia
import re
```

```python
#extract plain text content
wiki = wikipedia.page('Rajahmundry')
text = wiki.content
```

```python
text
```

```python
# clean text
text = re.sub(r'==.*?==+', '', text)
text = text.replace('\n', '')
```

```python
text
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_cloud(wordcloud):
    plt.figure(figsize = (12,8))
    plt.imshow(wordcloud)
    plt.axis('off')
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
wordcloud = WordCloud(width= 3000, height = 2000,
                      random_state=1, background_color='green',
                      colormap='terrain', collocations=False, stopwords = STOPWORDS).generate(text)
```

```python
plot_cloud(wordcloud)
```
